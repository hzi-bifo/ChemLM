```python
sns.set(style="ticks")
from matplotlib.pyplot import figure
from matplotlib.lines import Line2D

figure(figsize=(12,8))

g = sns.swarmplot(data=df, x="Models", y="F1-score", hue="Dataset",palette=sns.set_palette([ '#fdae61','#abd9e9','#2c7bb6']), alpha=0.6, s=17)
g.set_xticklabels(labels = ['MolBERT','MolFormer', 'ChemBERTa', 'GAT', 'MPNN', 'GCNN', 'ChemLM \nvanilla', 'ChemLM \ndomain-adapted',
	'ChemLM domain-adapted \n& optimized'], rotation=30, size=17)
plt.yticks(fontsize=17)
plt.xlabel('Models', fontsize=20)
plt.ylabel('F1-score',fontsize=20)

median_values = df.groupby("Models")["F1-score"].mean().reset_index()

sns.pointplot(data=median_values, x="Models", y="F1-score", color="#d7191c", label="Mean F1-score\n across datasets",markersize=15, markers="D",linestyles="")
#sns.lineplot(x = , y =  dat, color = '#fc8d62')
plt.legend( loc ="upper left", bbox_to_anchor=(1, 1), fontsize = 6, prop={'size': 12})
plt.ylim(0.0, 1)
plt.grid()  #just add this
plt.tight_layout()
plt.savefig('xmodels.png',dpi=300)
```
